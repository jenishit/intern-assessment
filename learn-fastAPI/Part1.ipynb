Introduction to FastAPI
FastAPI is a modern python framework for building APIs.

It is built on two important components:
1. Starlette : It handles web features like routing and async requests
2. Pydantic : It handles data validation and type checking

FastAPI also runs using ASGI server Uvicorn
ASGI server is a web server that implements the Asynchronous Server Gateway Interface, a standar for handling asynchronous Python web applciations and frameworks. It supports asynchronous programming, WebSockets, and real-time communication, unlike the synchronous WSGI.

Key Features of FastAPI:
a. High performance
b. Automatic data validation using Python type hints
c. Automatic interactive API docs
d. Built-in support for authentication
e. Async support using async/await
f. Easy integration with DBs

Basic FastAPI application

```python
from fastapi import FastAPI

app =FastAPI()

@app.get("/") #This is HTTP method GET with URL /
def read_root():
    return {"message": "Hello and Good morning!"}
```

HTTP Methods in FastAPI
Method             Purpose
GET                 Retrieve data
POST                Create data
PUT                 Update data
DELETE              Remove data

```python
@app.get("/users")
def get_users():
    return ["Ram", "Shyam", "Hanuman"]

@app.post("/users")
def create_user():
    return {"message": "User successfully created"}
```

Async vs Sync Functions
FastAPI supports asynchronous programming.
@app.get("/sync")
def sync_function():
    return {"message": "This is a normal function"}

@app.get("/async")
def async_function():
    return {"message": "This is an async fucntion"}

Async is useful for:
a. Database queries
b. External API calls
c. File processing

Path Parameters:
Path parameter is the part of the URL itself. Like -> /users/5
where 5 is the user id
Example:
@app.get("/users/{user_id}")
def get_user(user_id: int):
    return {"user_id": user_id}

Here {user_id} is a dynamic part and user_id: int tells FastAPI the type of variable user_id
if someone sends /users/abc
FastAPI returns
{
 "detail":[
  {
   "type":"int_parsing",
   "msg":"Input should be a valid integer"
  }
 ]
}

You can also set multiple path parameters:
@app.get("/users/{user_id}/orders/{order_id})
def get_order(user_id: int, order_id: int):
    return {
        "user_id": user_id,
        "order_id": order_id
    }

Query Parameters
It appears after ? in URL like -> /products?limit=10&skip=5
Example:
@app.get("/products")
def get_products(limit: int, skip: int):
    return {
        "limit": limit,
        "skip": skip
    }

Optional Query Parameters
from typing import Optional

@app.get("/items")
def read_items(q: Optional[str] = None):
    return {"query": q}

Default Query Paramaters
@app.get("/products")
def get_products(limit: int = 10):
    return {"limit": limit}
